```python
from sklearn.preprocessing import MinMaxScaler
df = pd.DataFrame(df)

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Normalize all features except the target variable (Age)
df= df.copy()
df[["Respondent's Oral","Blood Insulin Levels"]] = scaler.fit_transform(df[["Respondent's Oral",	"Blood Insulin Levels"]])
print("Normalized Data:\n", df)
```

```python
df.head()
```

```python
def wisker(df):
  Q1 = df.quantile(0.25)
  Q3 = df.quantile(0.75)
  IQR = Q3 - Q1
  lw=Q1-1.5*IQR
  uw=Q3+1.5*IQR
  return lw,uw
```

```python
wisker(df["Body Mass Index"])
```

```python
wisker(df["Blood Glucose after fasting"])
```

```python
wisker(df["Blood Insulin Levels"])
```

```python
wisker(df["Respondent's Oral"])
```

```python
for i in ["Body Mass Index","Blood Glucose after fasting","Blood Insulin Levels","Respondent's Oral"]:
  lw,uw=wisker(df[i])
  df[i]=np.where(df[i]<lw,lw,df[i])
  df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators=1500,max_depth=5,min_samples_split=6,min_samples_leaf=4,random_state=42)
rf_model.fit(x_train,y_train)
train_rf_model = rf_model.predict(x_train)
rf_model_predictions = rf_model.predict(x_test)

RF_R2_Score = r2_score(y_test,rf_model_predictions)

print('RMSE :' , round(np.sqrt(mean_squared_error(y_test,rf_model_predictions)),2))
print('R2 Score : ', RF_R2_Score)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    print(f"Model: {model.__class__.__name__}")
    print(f"R² Score: {r2:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print("-" * 50)

# Assume your models are trained as:
models = {
    "RandomForest": rf_model,
    "GradientBoosting": GBM_model,
    "XGBoost": XGB_model,
    "LinearRegression":model
}

# Evaluate each model
for name, model in models.items():
    evaluate_model(model, x_test, y_test)
```
